# Rumor Propagation Model: Daley-Kendall Approach

## Introduction

We study how a rumor spreads through a fixed-size population by using a simple compartmental model.  
Individuals are divided into three groups:

- **Ignorants (I):** individuals who have not heard the rumor.
- **Spreaders (S):** individuals who are actively spreading the rumor.
- **Stiflers (R):** individuals who have heard the rumor but no longer spread it.

We are interested in modeling how the fractions of Ignorants, Spreaders, and Stiflers change over time.

---

## Model Structure

The structure of the model is based on how individuals transition between states:

- **Ignorants become Spreaders** when they hear the rumor from an active Spreader.
- **Spreaders become Stiflers** when they interact with another Spreader or a Stifler and realize the rumor is already well known.

## Variables

We define the following time-dependent variables:

- \( I(t) \): fraction of individuals who are Ignorants at time \( t \).
- \( S(t) \): fraction of individuals who are Spreaders at time \( t \).
- \( R(t) \): fraction of individuals who are Stiflers at time \( t \).

The total population is assumed constant:

\[
I(t) + S(t) + R(t) = 1, \quad \forall t
\]

---

## Model Assumptions

- The population is well-mixed (any individual can interact with any other).
- There are no births, deaths, or migrations (closed system).
- Individuals interact in continuous time.
- Upon interaction, transitions happen immediately.

---

## Mathematical Model

The system of differential equations describing the dynamics is:


\begin{aligned}
\frac{dI}{dt} &= -\beta I S \\
\frac{dS}{dt} &= \beta I S - \gamma S (S + R) \\
\frac{dR}{dt} &= \gamma S (S + R)
\end{aligned}


where:

- $ \beta > 0 $ is the **spreading rate** (how easily the rumor is transmitted),
- $ \gamma > 0 $ is the **stifling rate** (how easily spreaders become stiflers upon meeting someone else who knows the rumor).

---

## Interpretation of the Equations

- The number of Ignorants decreases as they hear the rumor from Spreaders.
- Spreaders increase by recruiting Ignorants and decrease by becoming Stiflers.
- Stiflers increase as Spreaders lose interest upon realizing others already know the rumor.
